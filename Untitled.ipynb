```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import spearmanr, pointbiserialr
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import StandardScaler
import statsmodels.formula.api as smf
import statsmodels.api as sm
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-whitegrid')

cols = ['path_efficiency', 'movement_smoothness', 'idle_time_pct',
        'tool_switching_per_min', 'site_visibility_pct',
        'two_tools_pct', 'speed', 'workspace_coverage']
# dropped composite_score itself
# Load data
outcomes  = pd.read_excel("Copy of ds_take_home_clinical_outcomes.csv.xlsx")
metrics = pd.read_excel("Copy of ds_take_home_procedure_metrics.csv.xlsx")

# Merge on procedure_id — keeps only rows present in both files (120 rows)
df = metrics.merge(outcomes, on='procedure_id', how='inner')

print(f'Metrics  : {metrics.shape}')
print(f'Outcomes : {outcomes.shape}')
print(f'Merged   : {df.shape}')
```

```python
print('=== METRICS ===')

display(metrics.describe().round(2))
display(outcomes.describe().round(2))
```

```python
# print(f'{"Metric":<30} {"r":>6}  {"p-value":>8}  {"Sig":>4}')
# print('-' * 55)

corr_results = {}
for col in cols:
    r, p = stats.pearsonr(metrics[col], metrics['composite_score'])
    sig = '***' if p < 0.001 else '**' if p < 0.01 else '*' if p < 0.05 else 'ns'
    corr_results[col] = (r, p)
    print(f'{col:<30} {r:>+6.3f}  {p:>8.4f}  {sig:>4}')

# print(corr_results)
```

```python
# scatter each metric against the score
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
axes = axes.flatten()

for i, col in enumerate(cols):
    r, p = corr_results[col]
    axes[i].scatter(metrics[col], metrics['composite_score'], alpha=0.3, s=20, color='steelblue')
    m, b = np.polyfit(metrics[col], metrics['composite_score'], 1)
    x_line = np.linspace(metrics[col].min(), metrics[col].max(), 100)
    axes[i].plot(x_line, m * x_line + b, color='red', linewidth=1.5)
    axes[i].set_xlabel(col, fontsize=9)
    axes[i].set_ylabel('Composite Score', fontsize=9)
    axes[i].set_title(f'r = {r:.3f}  (p={p:.4f})', fontsize=9,
                      color='darkgreen' if p < 0.05 else 'gray')

plt.suptitle('Individual Metrics vs Composite Score (n=300)', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Step 3: R² — how much of the score do all 8 metrics explain together?
X   = sm.add_constant(metrics[cols])
ols = sm.OLS(metrics['composite_score'], X).fit()

print(f'R²          = {ols.rsquared:.3f}')
print(f'Adjusted R² = {ols.rsquared_adj:.3f}')
print()
print(f'The 8 metrics together explain {ols.rsquared:.1%} of composite score variance.')
print(f'R2 = {ols.rsquared:.3f}  — so ~{1-ols.rsquared:.0%} is something else in the formula')
```

```python
# --- Step 4: Standardized beta coefficients — relative weight of each metric ---
# Standardize so all metrics are on equal footing (mean=0, std=1)
scaler   = StandardScaler()
X_scaled = scaler.fit_transform(metrics[cols])
ols_std  = sm.OLS(metrics['composite_score'], sm.add_constant(X_scaled)).fit()

betas       = list(zip(cols, ols_std.params[1:]))
betas_sorted = sorted(betas, key=lambda x: abs(x[1]), reverse=True)
total_weight = sum(abs(b) for _, b in betas_sorted)

# print(f'{"Metric":<30} {"Beta":>8}  {"Weight %":>9}')
# print('-' * 52)
for col, beta in betas_sorted:
    pct = abs(beta) / total_weight * 100
    print(f'{col:<30}  {beta:>+.3f}  ({pct:.0f}%)')
# path_efficiency way heavier than the others  score isn't balanced at all
```

```python
# --- Outcome 1: Complication ---
print('OUTCOME 1: Complication (Yes/No)')

comp_yes = df[df['complication'] == 1]['composite_score']
comp_no  = df[df['complication'] == 0]['composite_score']

print(f'No complication (n={len(comp_no)}): mean={comp_no.mean():.1f}, std={comp_no.std():.1f}')
print(f'Complication    (n={len(comp_yes)}): mean={comp_yes.mean():.1f}, std={comp_yes.std():.1f}')

# Mann-Whitney U — does not assume normality, robust for small/unequal groups
u_stat, p_mwu = stats.mannwhitneyu(comp_yes, comp_no, alternative='two-sided')
print(f'\nMann-Whitney U: p = {p_mwu:.3f}  {"Significant" if p_mwu < 0.05 else "NOT significant"}')

# AUC — how well does the score discriminate between groups?
auc = roc_auc_score(df['complication'], df['composite_score'])
print(f'AUC = {auc:.3f}  (0.5=random, 1.0=perfect)')

# Logistic regression
log_unadj = smf.logit('complication ~ composite_score', data=df).fit(disp=0)
log_adj   = smf.logit('complication ~ composite_score + procedure_length_min + C(surgeon_id)', data=df).fit(disp=0)

print(f'\nLogistic Regression:')
print(f'  Unadjusted: OR={np.exp(log_unadj.params["composite_score"]):.3f}, p={log_unadj.pvalues["composite_score"]:.3f}')
print(f'  Adjusted:   OR={np.exp(log_adj.params["composite_score"]):.3f}, p={log_adj.pvalues["composite_score"]:.3f}')
print()
print('Note: Adjusted OR collapses to ~1.0 — surgeon identity was the real driver.')
```

```python
# --- Outcome 2: Complete Resection ---
print('OUTCOME 2: Complete Resection (Yes/No)')

df_res  = df[df['complete_resection'].notna()].copy()
res_yes = df_res[df_res['complete_resection'] == 1]['composite_score']
res_no  = df_res[df_res['complete_resection'] == 0]['composite_score']

print(f'Applicable cases: {len(df_res)}  (38 NaN excluded — not applicable)')
print(f'Complete   (n={len(res_yes)}): mean={res_yes.mean():.1f}, std={res_yes.std():.1f}')
print(f'Incomplete (n={len(res_no)}): mean={res_no.mean():.1f}, std={res_no.std():.1f}')

u2, p_mwu2 = stats.mannwhitneyu(res_yes, res_no, alternative='two-sided')
print(f'\nMann-Whitney U: p = {p_mwu2:.3f}  → {"Significant" if p_mwu2 < 0.05 else "NOT significant"}')

# Unadjusted + partially adjusted (cannot add all 10 surgeon dummies with only 11 incomplete cases)
log_res_u = smf.logit('complete_resection ~ composite_score', data=df_res).fit(disp=0)
log_res_a = smf.logit('complete_resection ~ composite_score + procedure_length_min', data=df_res).fit(disp=0)

print(f'\nLogistic Regression:')
print(f'  Unadjusted:                       OR={np.exp(log_res_u.params["composite_score"]):.3f}, p={log_res_u.pvalues["composite_score"]:.3f}')
print(f'  Adjusted (score + length only):   OR={np.exp(log_res_a.params["composite_score"]):.3f}, p={log_res_a.pvalues["composite_score"]:.3f}')
print()
# print('Note: Cannot adjust for all 10 surgeons — only 11 incomplete resection cases total.')
# print('Too few events to support that many parameters (singular matrix error).')
# print('This itself is a limitation to report.')
```

```python
# --- Outcome 3: Hospital Stay ---
print('OUTCOME 3: Hospital Stay (days)')

r_hosp,   p_hosp   = stats.pearsonr(df['composite_score'], df['hospital_stay_days'])
rho_hosp, p_hosp_s = spearmanr(df['composite_score'], df['hospital_stay_days'])

print(f'Pearson  r = {r_hosp:.3f}, p = {p_hosp:.3f}  {"Significant" if p_hosp < 0.05 else "NOT significant"}')
print(f'Spearman ρ = {rho_hosp:.3f}, p = {p_hosp_s:.3f}  {"Significant" if p_hosp_s < 0.05 else "NOT significant"}')

ols_h_u = smf.ols('hospital_stay_days ~ composite_score', data=df).fit()
ols_h_a = smf.ols('hospital_stay_days ~ composite_score + procedure_length_min + C(surgeon_id)', data=df).fit()

print(f'\nLinear Regression:')
print(f'  Unadjusted: coef={ols_h_u.params["composite_score"]:.4f}, p={ols_h_u.pvalues["composite_score"]:.3f}')
print(f'  Adjusted:   coef={ols_h_a.params["composite_score"]:.4f}, p={ols_h_a.pvalues["composite_score"]:.3f}')
print(f'\nPractical effect: +10 points in score {ols_h_u.params["composite_score"] * 10:.2f} days change (unadjusted)')
print('Negligible clinical impact.')
```

```python
# --- Part 1B Summary Plots ---
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot 1: Complication
axes[0].boxplot(
    [comp_no, comp_yes],
    labels=[f'No Complication\n(n={len(comp_no)})', f'Complication\n(n={len(comp_yes)})'],
    patch_artist=True,
    boxprops=dict(facecolor='lightblue'),
    medianprops=dict(color='red', linewidth=2)
)
axes[0].set_ylabel('Composite Score')
axes[0].set_title(f'Score vs Complication\n(Mann-Whitney p={p_mwu:.3f})')

# Plot 2: Complete Resection
axes[1].boxplot(
    [res_yes, res_no],
    labels=[f'Complete\n(n={len(res_yes)})', f'Incomplete\n(n={len(res_no)})'],
    patch_artist=True,
    boxprops=dict(facecolor='lightgreen'),
    medianprops=dict(color='red', linewidth=2)
)
axes[1].set_ylabel('Composite Score')
axes[1].set_title(f'Score vs Resection\n(Mann-Whitney p={p_mwu2:.3f})')

# Plot 3: Hospital Stay scatter
axes[2].scatter(df['composite_score'], df['hospital_stay_days'], alpha=0.4, color='steelblue', s=25)
m, b = np.polyfit(df['composite_score'], df['hospital_stay_days'], 1)
x_line = np.linspace(df['composite_score'].min(), df['composite_score'].max(), 100)
axes[2].plot(x_line, m * x_line + b, color='red', linewidth=1.5)
axes[2].set_xlabel('Composite Score')
axes[2].set_ylabel('Hospital Stay (days)')
axes[2].set_title(f'Score vs Hospital Stay\n(r={r_hosp:.3f}, p={p_hosp:.3f})')

plt.suptitle('Composite Score vs Clinical Outcomes', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# --- Surgeon variability overview ---
surg_stats = metrics.groupby('surgeon_id').agg(
    n=('composite_score', 'count'),
    mean_score=('composite_score', 'mean'),
    std_score=('composite_score', 'std')
).sort_values('mean_score', ascending=False)

print('Surgeon-level summary:')
print(surg_stats.round(2).to_string())

# Intraclass Correlation Coefficient (ICC)
# ICC tells us: what % of score variance is due to surgeon vs individual procedure?
grand_mean    = metrics['composite_score'].mean()
between_var   = surg_stats['mean_score'].var()
within_var    = metrics.groupby('surgeon_id')['composite_score'].var().mean()
icc           = between_var / (between_var + within_var)

print(f'\nIntraclass Correlation (ICC) = {icc:.3f}')
print(f'{icc:.1%} of score variance is explained by which surgeon performed the procedure.')
print('This justifies using a mixed effects model.')
```

```python
# --- Surgeon score distribution plot ---
fig, ax = plt.subplots(figsize=(12, 5))
surgeon_order = surg_stats.index.tolist()

data_by_surgeon = [metrics[metrics['surgeon_id'] == s]['composite_score'].values for s in surgeon_order]
bp = ax.boxplot(data_by_surgeon, labels=surgeon_order, patch_artist=True)

colors = plt.cm.Set3(np.linspace(0, 1, len(surgeon_order)))
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

ax.axhline(grand_mean, color='red', linestyle='--', linewidth=1.5, label=f'Grand mean = {grand_mean:.1f}')
ax.set_xlabel('Surgeon ID')
ax.set_ylabel('Composite Score')
ax.set_title('Composite Score Distribution by Surgeon')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# --- Mixed Effects Model: hospital stay ~ score (surgeon as random effect) ---
# Using hospital_stay as outcome (continuous — most stable for mixed effects with this sample size)
mixed_model = smf.mixedlm(
    'hospital_stay_days ~ composite_score + procedure_length_min',
    data=df,
    groups=df['surgeon_id']   # surgeon = random effect
).fit(reml=True)

coef = mixed_model.params['composite_score']
p    = mixed_model.pvalues['composite_score']

print('Mixed Effects Model: hospital_stay ~ score + length + (1|surgeon)')
print(f'  composite_score coef = {coef:.4f}')
print(f'  p-value              = {p:.3f}')
print(f'  Random effect (surgeon variance) = {mixed_model.cov_re.values[0][0]:.4f}')
print()
print('Compare to fixed effects OLS (unadjusted):')
print(f'  OLS coef = {ols_h_u.params["composite_score"]:.4f}, p = {ols_h_u.pvalues["composite_score"]:.3f}')
print()
print('Interpretation: after properly accounting for surgeon clustering,')
print('the composite score still does not significantly predict hospital stay.')
```

```python
from statsmodels.stats.multitest import multipletests

results = []

for col in cols:
    # Complication (binary) — AUC + logistic p-value
    auc_comp  = roc_auc_score(df['complication'], df[col])
    log_comp  = smf.logit(f'complication ~ {col}', data=df).fit(disp=0)
    p_comp    = log_comp.pvalues[col]

    # Complete resection (binary)
    auc_res   = roc_auc_score(df_res['complete_resection'], df_res[col])
    log_res   = smf.logit(f'complete_resection ~ {col}', data=df_res).fit(disp=0)
    p_res     = log_res.pvalues[col]

    # Hospital stay (continuous) — Pearson r
    r_stay, p_stay = stats.pearsonr(df[col], df['hospital_stay_days'])

    results.append({
        'metric'         : col,
        'auc_complication': auc_comp,
        'p_complication' : p_comp,
        'auc_resection'  : auc_res,
        'p_resection'    : p_res,
        'r_hospital_stay': r_stay,
        'p_hospital_stay': p_stay
    })

results_df = pd.DataFrame(results)

# FDR correction across all p-values (24 tests)
all_p = results_df[['p_complication','p_resection','p_hospital_stay']].values.flatten()
_, p_adj, _, _ = multipletests(all_p, method='fdr_bh')
results_df['p_comp_adj']  = p_adj[0:8]
results_df['p_res_adj']   = p_adj[8:16]
results_df['p_stay_adj']  = p_adj[16:24]

# Add composite score row for comparison
composite_row = {
    'metric'          : 'COMPOSITE SCORE',
    'auc_complication': roc_auc_score(df['complication'], df['composite_score']),
    'p_complication'  : log_unadj.pvalues['composite_score'],
    'auc_resection'   : roc_auc_score(df_res['complete_resection'], df_res['composite_score']),
    'p_resection'     : log_res_u.pvalues['composite_score'],
    'r_hospital_stay' : r_hosp,
    'p_hospital_stay' : p_hosp
}
results_df = pd.concat([results_df, pd.DataFrame([composite_row])], ignore_index=True)

display(results_df[['metric','auc_complication','auc_resection','r_hospital_stay']].round(3))
```

```python
# --- Heatmap: AUC and correlation of each metric vs each outcome ---
plot_df = results_df.set_index('metric')[['auc_complication','auc_resection','r_hospital_stay']]
plot_df.columns = ['AUC: Complication','AUC: Resection','r: Hospital Stay']

fig, ax = plt.subplots(figsize=(10, 6))
sns.heatmap(
    plot_df.astype(float),
    annot=True, fmt='.3f', cmap='RdYlGn',
    center=0.5, vmin=0.3, vmax=0.8,
    linewidths=0.5, ax=ax
)
ax.set_title('Predictive Performance: Individual Metrics vs Composite Score\n(AUC for binary outcomes, Pearson r for hospital stay)',
             fontsize=11, fontweight='bold')
plt.tight_layout()
plt.show()

# Best metric per outcome
print('Best individual metric per outcome:')
non_composite = results_df[results_df['metric'] != 'COMPOSITE SCORE']
print(f'  Complication  : {non_composite.loc[non_composite["auc_complication"].idxmax(), "metric"]}  (AUC={non_composite["auc_complication"].max():.3f})')
print(f'  Resection     : {non_composite.loc[non_composite["auc_resection"].idxmax(), "metric"]}  (AUC={non_composite["auc_resection"].max():.3f})')
print(f'  Hospital Stay : {non_composite.loc[non_composite["r_hospital_stay"].abs().idxmax(), "metric"]}  (r={non_composite["r_hospital_stay"].abs().max():.3f})')
```
